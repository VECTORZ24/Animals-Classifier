```python
#importing libraries
from imutils import paths
```

```python
imagePaths=list(paths.list_images(r'C:\Users\user\Desktop\animal_database'))
imagePaths[0].split('\\')[-3]
```

```python
#targeting animal names
def getClassNames():
    animalNames=[]
    for imagePath in imagePaths:
        animal=imagePath.split('\\')[-3]
        if(animal not in animalNames):
            animalNames.append(animal)
    return animalNames
getClassNames()
```

```python
labels=getClassNames()
```
